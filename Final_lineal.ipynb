# Задание 6
Найдите сумму сингулярных чисел для матрицы:

$A=\left(\begin{array}{ccc}-2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1\end{array}\right)$

Для нахождения суммы сингулярных чисел матрицы $A$, необходимо вычислить собственные значения матрицы $A^T A$, а затем взять их квадратные корни

1) Вычислим произведение $A^T A$:

$$
A^T A = \begin{pmatrix}
-2 & 0 & 1 \\
0 & -2 & 1 \\
1 & 1 & -1
\end{pmatrix}
\begin{pmatrix}
-2 & 0 & 1 \\
0 & -2 & 1 \\
1 & 1 & -1
\end{pmatrix}
=
\begin{pmatrix}
4 + 0 + 1 & 0 + 0 + 1 & -2 + 0 - 1 \\
0 + 0 + 1 & 0 + 4 + 1 & 0 - 2 - 1 \\
-2 + 0 - 1 & 0 - 2 - 1 & 1 + 1 + 1
\end{pmatrix}
=
\begin{pmatrix}
5 & 1 & -3 \\
1 & 5 & -3 \\
-3 & -3 & 3
\end{pmatrix}
$$

2) Найдем собственные значения матрицы $A^T A$

    Для нахождения собственных значений, решим характеристическое уравнение:

    $
    \det(A^T A - \lambda I) = 0
    $

    Вычислим детерминант:

    $$
    \det(A^T A - \lambda I) = \det \begin{pmatrix}
    5 - \lambda & 1 & -3 \\
    1 & 5 - \lambda & -3 \\
    -3 & -3 & 3 - \lambda
    \end{pmatrix}
    =
    (5 - \lambda) \det \begin{pmatrix} 5 - \lambda & -3 \\ -3 & 3 - \lambda \end{pmatrix} - 1 \det \begin{pmatrix} 1 & -3 \\ -3 & 3 - \lambda \end{pmatrix} + (-3) \det \begin{pmatrix} 1 & 5 - \lambda \\ -3 & -3 \end{pmatrix}
    =
    -\lambda^3 + 13\lambda^2-36\lambda=-\lambda(\lambda^2-13\lambda+36) = -\lambda(\lambda-4)(\lambda-9)=0
    $$
    Следовательно:
    $
    \lambda_1 = 9,\lambda_2 = 4,\lambda_3 = 0
    $

3) Найдем сингулярные числа:

    Сингулярные числа это квадратные корни из собственных значений матрицы $A^T A$:

    $
    \sigma_1 = \sqrt{\lambda_1} = \sqrt{9} = 3
    $

    $
    \sigma_2 = \sqrt{\lambda_2} = \sqrt{4} = 2
    $

    $
    \sigma_3 = \sqrt{\lambda_3} = \sqrt{0} = 0
    $

4) Найдем сумму сингулярных чисел

    $$
    \sigma_1 + \sigma_2 + \sigma_3 = 3 + 2 + 0 = 5
    $$

# Задание 10
Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
# Предикторы
X = data.drop(columns=["cnt"])

# Целевая переменная
y = data["cnt"]

# Добавляем единичный столбец
X = np.c_[np.ones(X.shape[0]), X]

# Преобразуем в numpy array для дальнейшей работы
X = np.array(X)
y = np.array(y)

# Находим коэффициенты линейной регрессии
beta = np.linalg.inv(X.T @ X) @ X.T @ y

# Формируем предсказание модели
y_pred = X @ beta

# Оцениваем метрики качества
mse = np.mean((y - y_pred) ** 2)
mape = np.mean(np.abs((y - y_pred) / y)) * 100

print(f'mse: {mse} \nmape: {mape}')
```

Модель имеет высокие ошибки, что говорит о том, что ее предсказания далеко от реальных значений
